# Blanko Diagramme mit zuf√§lligen Daten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 10
pd.DataFrame(np.random.exponential(size=(20,1)), columns=['Column']).sort_values('Column', ascending=False).reset_index(drop=True).head(n+1).tail(n).plot(figsize=(10,5), kind='bar', legend=False, title='√úberschrift')
plt.xticks(rotation=70)
plt.xlabel('Gruppierung')
plt.ylabel('Anzahl oder Wert')
plt.tight_layout()
plt.savefig('Grafiken/00_Blanko_Balken.pdf')
plt.show()
```

```python
n = 10
pd.DataFrame(np.random.exponential(size=(20,2)), columns=['Spiel 1', 'Spiel 2']).reset_index(drop=True).head(n+1).tail(n).plot(figsize=(10,5), title='√úberschrift')
plt.xticks(rotation=70)
plt.xlabel('Datum')
plt.ylabel('Anzahl oder Wert')
plt.tight_layout()
plt.savefig('Grafiken/00_Blanko_Evolution.pdf')
plt.show()
```

```python
from matplotlib import cm
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from matplotlib import patches
import cartopy.io.shapereader as shpreader
import itertools
import numpy as np

shapename = 'admin_0_countries'
countries_shp = shpreader.natural_earth(resolution='110m',
                                        category='cultural', name=shapename)

plt.figure(figsize=(14,10))
ax = plt.axes(projection=ccrs.PlateCarree())
for country in shpreader.Reader(countries_shp).records():
    ax.add_geometries(country.geometry, ccrs.PlateCarree(),
                      facecolor=cm.tab10(np.random.randint(5)),
                      label=country.attributes['NAME_LONG'])

plt.title('√úberschrift')
plt.tight_layout()
plt.savefig('Grafiken/00_Blanko_Welt.pdf')
plt.show()
```
