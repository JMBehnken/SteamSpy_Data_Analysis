# Compute the correlation for the 2weeks playtime

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('RPG_Games_Data.csv', index_col=0).drop_duplicates()[['average_2weeks', 'name']]
print(df.shape)
df.head(2)
```

```python
df.columns
```

```python
df_corr = df.groupby([df.index, 'name']).mean().reset_index().pivot(index='date', columns='name', values='average_2weeks').fillna(0).corr()
df_corr.head()
```

# Which games correlate?

```python
game = ['The Elder Scrolls V: Skyrim', 'Stardew Valley', 'The Witcher 3: Wild Hunt'][0]
df_corr[game].sort_values(ascending=False).head(10)
```
