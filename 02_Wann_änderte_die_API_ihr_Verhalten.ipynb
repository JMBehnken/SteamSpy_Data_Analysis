# Wann änderte die API ihr Verhalten?

Irgendwann änderte die SteamSpy-API ihr Verhalten in die Richtung, dass die Spalten 'CCU' und 'Tags' nicht mit mitgesendet wurden. Hinzu kamen die Spalten 'Positiv' und 'Negativ'.

```python
# Import
import os
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Path to the data
data_path = '/home/jm/Desktop/Gamestar_Data/API/'

# Paths for all datafiles
data_files= sorted([os.path.join(data_path, file) for file in os.listdir(data_path)])

# Extract date and filesize
data = [[datetime.strptime(file.rsplit('/', 1)[-1].split('_', 1)[0], '%Y-%m-%d'), os.stat(file).st_size/1000] for file in data_files]
```

```python
# Parse the data to a DataFrame
df = pd.DataFrame(data, columns=['Date', 'Size']).set_index('Date')
```

```python
# Plot the filesize by day
df.plot(title='Filesize by day', grid=True, figsize=(14,4))
plt.xlabel('Date')
plt.ylabel('Filesize in kb')
plt.show()
```

Am 25.11. kamen die beiden Spalten 'Positiv' und 'Negativ' hinzu und am 29.11. vielen die 'Tags' und 'CCU' weg.
